```python
"""
Example 2: LoRA Configuration

This example demonstrates how to add LoRA adapters to a model,
showing compatibility with Unsloth's get_peft_model API.
"""

from mlx_tune import FastLanguageModel

# def main():
print("= - 02_lora_conf.ipynb:11" * 60)
print("MLXTune Example: LoRA Configuration - 02_lora_conf.ipynb:12")
print("= - 02_lora_conf.ipynb:13" * 60)

# Load a model
print("\n1. Loading model... - 02_lora_conf.ipynb:16")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="mlx-community/Llama-3.2-1B-Instruct-4bit",
    max_seq_length=2048,
    load_in_4bit=True,
)
print("✓ Model loaded! - 02_lora_conf.ipynb:22")




# if __name__ == "__main__":
    # main()
```

```python
# Add LoRA adapters
print("\n2. Adding LoRA adapters... - 02_lora_conf.ipynb:2")
model = FastLanguageModel.get_peft_model(
    model,
    r=16,  # LoRA rank
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ],
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=3407,
)
print("✓ LoRA adapters configured! - 02_lora_conf.ipynb:16")

# Check LoRA configuration
print("\n3. LoRA Configuration: - 02_lora_conf.ipynb:19")
if hasattr(model, 'lora_config') and model.lora_config:
    print(f"LoRA Rank: {model.lora_config['r']} - 02_lora_conf.ipynb:21")
    print(f"LoRA Alpha: {model.lora_config['lora_alpha']} - 02_lora_conf.ipynb:22")
    print(f"LoRA Dropout: {model.lora_config['lora_dropout']} - 02_lora_conf.ipynb:23")
    print(f"Target Modules: {model.lora_config['target_modules']} - 02_lora_conf.ipynb:24")
    print(f"Bias: {model.lora_config['bias']} - 02_lora_conf.ipynb:25")
    print(f"LoRA Enabled: {model.lora_enabled} - 02_lora_conf.ipynb:26")
else:
    print("LoRA configuration not found - 02_lora_conf.ipynb:28")

print("\n - 02_lora_conf.ipynb:30" + "=" * 60)
print("Example completed successfully! - 02_lora_conf.ipynb:31")
print("Ready for finetuning! - 02_lora_conf.ipynb:32")
print("= - 02_lora_conf.ipynb:33" * 60)
```
